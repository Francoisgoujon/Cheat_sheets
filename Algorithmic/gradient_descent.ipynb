# Cheat Sheet - Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import norm
```

```python
def grad_des(f_grad, step_size = 0.1, eps=10e-3, max_iter = 1000):
    x = 0
    path = [x]
    for iter in range(max_iter):
        g = f_grad(x)
        if norm(g) < eps:
            break
        x -= step_size*g
        path.append(x)
    print(f"Optimum found in {iter} iterations")
    return x, np.array(path)

def grad_des_opt(f, f_grad, eps=10e-3, max_iter = 1000):
    x = 0
    path = [x]
    steps = np.geomspace(0.001, 1, 30)
    st_pa = []
    for iter in range(max_iter):
        g = f_grad(x)
        if norm(g) < eps:
            break
        step = steps[np.argmin(f(x - steps*g))]
        st_pa.append(step)
        x -= step*g
        path.append(x)
    print(f"Optimum found in {iter} iterations")
    return x, np.array(path), st_pa
```

```python
def para(x):
    return (x-5)**2
def grad_para(x):
    return 2*(x-5)
```

```python
xs = np.arange(-3, 10, 0.1)
ys = para(xs)
```

```python
x0, pa = grad_des(grad_para)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(xs, ys)
plt.scatter(pa, para(pa))
plt.show()
```

```python
x0, pa, sts = grad_des_opt(para, grad_para)
```
